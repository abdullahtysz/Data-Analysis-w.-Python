```python
x = np.arange(1,10)
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(10,4))
ax[0].plot(x,x**2,x,x**3,x,x**4)
ax[1].plot(x,x**2,x,x**3,x,x**4)
ax[1].set_xlim([2,4]) # x ekseninde 2,4 arasna zoom
ax[1].set_ylim([9,100]) #y ekseninde 9,100 arasna zoom
```
