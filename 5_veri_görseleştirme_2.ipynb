```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset("titanic")
df.head()
```

```python
# 1) barplot
```

```python
# Cinsiyete göre hayatta kalma oranını grafik olarak göster ve barplot grafiğini kullan
sns.barplot(x="sex",y="survived",data=df);  # data=df datayı df den al demek
```

```python
# Class a göre hayatta kalma oranlarını getir, ama eksenleri değiştir
sns.barplot(x="pclass",y="survived",data=df)
```

```python
sns.barplot(x="sex",y="survived",hue="pclass",data=df)
```

```python
tips=sns.load_dataset("tips")
df=tips.copy()
df.head()
```

```python
df["sex"].value_counts()
```

```python
df["smoker"].value_counts()
```

```python
sns.barplot(x="sex",y="total_bill",data=df);
# erkekler bayanlardan daha fazla harcama yapıyor
```

```python
sns.barplot(x="sex",y="total_bill",hue="time",data=df);
```

```python
sns.barplot(x="sex",y="tip",data=df)
```

```python
sns.barplot(x="sex",y="tip", hue="smoker", data=df)
```

```python
sns.barplot(x="day",y="total_bill",hue="time",data=df);
```
